```python
# Date columns
zulo_dates = zulo_bank[['StartDate','EndDate','OpeningDate','TransactionDate']]
```

```python
zulo_bank[['StartDate','EndDate','OpeningDate','TransactionDate']] = zulo_dates.apply(pd.to_datetime)
```

```python
# Fill up missing values with approprite parameters
zulo_bank.fillna({
        'LoanAmount' : 0.0, 
        'InterestRate':'0.0'
}, inplace=True)
```

```python
zulo_bank.info()
```

```python
# Convert to 1NF
# Spliting the 'Fullname' into two parts first_name and last_name
# NB: The [[]] are used in pandas to select multiple column 
# str.split() the split strings into separate columns. 
# When expand=True is used, the result is a DataFrame where each part of the split string occupies its own column.

zulo_bank[['first_name', 'last_name']] = zulo_bank['FullName'].str.split(' ', n=1, expand=True)
zulo_bank.head()
```

# MODELLING THE DATABASE

### Convert from 1NF to 2NF

```python
# Convert from 1NF to 2NF

# Create Customer_table, remove duplicated values and reet the index values to ensure they are correctly numbered
customer =zulo_bank[['first_name','last_name','Email','Phone']].copy().drop_duplicates().reset_index(drop=True)
```

```python

# Create the customer_id column which will serve as the primary key.
customer['customer_id'] = range(1, len(customer)+1)
customer['customer_id']
```

```python
customer = customer[['customer_id','first_name','last_name','Email','Phone']]
customer
```

### Convert from 2NF to 3NF

Creating the Account_table

```python
# Creating the Account_table
account = zulo_bank[['AccountType','Balance','OpeningDate']].copy().drop_duplicates().reset_index(drop=True)
account['account_id'] = range(1, len(account) + 1)
account = account[['account_id','AccountType','Balance','OpeningDate']]
```

```python
account
```

### Creating the transcation Table

```python
transcations = zulo_bank[['TransactionType','Amount','TransactionDate']].copy().drop_duplicates().reset_index(drop=True)
transcations['transcations_id'] = range(1, len(transcations)+1)
# Arrange the colums of the table in the preffered order with transcation_ID in the 1st column
transcations = transcations[['transcations_id','TransactionType','Amount','TransactionDate']]
transcations
```

```python
zulo_bank.head(1)
```

### Creating the Loan Table

```python
loan = zulo_bank[['LoanType','LoanAmount','InterestRate','StartDate','EndDate']].copy().drop_duplicates().reset_index(drop=True)
loan['loan_id'] = range(1, len(loan)+1)
loan = loan[['loan_id','LoanType','LoanAmount','InterestRate','StartDate','EndDate']]
loan
```

### Merging all the dimension table to the zulo_bank dataframe and 
### slicing out the required column to form the database fact table

```python
# Merge operations to create the Zulo_Bank
# Merge zulo_bank with customer data based on name, email, and phone
zulo_bank = zulo_bank.merge(
    customer, 
    on=['first_name', 'last_name', 'Email', 'Phone'], 
    how='left'
)

# Merge the result with account data based on account type, balance, and opening date
zulo_bank = zulo_bank.merge(
    account, 
    on=['AccountType', 'Balance', 'OpeningDate'], 
    how='left'
)

# Merge the result with transaction data based on transaction type, amount, and transaction date
zulo_bank = zulo_bank.merge(
    transcations, 
    on=['TransactionType', 'Amount', 'TransactionDate'], 
    how='left'
)

# Merge the result with loan data based on loan type, loan amount, interest rate, start date, and end date
zulo_bank = zulo_bank.merge(
    loan, 
    on=['LoanType', 'LoanAmount', 'InterestRate', 'StartDate', 'EndDate'], 
    how='left'
)

zulo_bank                 
```

### zulo_bank database fact table

```python
# Slice the required columns from the merged zulo_bank DataFrame
zulo_bank_fact = zulo_bank[['customer_id', 'account_id', 'transcations_id', 'loan_id']]
zulo_bank_fact
```

## Creating the date table

```python
# Convert from the second Normal form to third normal form
# Create the date dimension table
# Define the start and end dates

start_date = datetime(2020,1,1)
current_date = datetime(2090,12,31)

# Calculate the number of days between start and current dates

num_days = (current_date - start_date).days
num_days
```

Creating the date table using dictionary with Date_ID Mapping

```python
# Create a dictionary with two keys: 'date_id' and 'date
# 'date_id': A list of sequential IDs starting from 1 up to the length of date_list
# 'date': The list of dates (datetime objects or date objects)
date_list = [start_date + timedelta(days=x) for x  in range(num_days + 1)]
date_list
date = {'date_id':[x for x in range(1,len(date_list)+ 1)], 'date':date_list}
#date['date']
```

### Date Dimension Table

```python
#create dataframe
# Create a DataFrame from the 'date' data
date_dim = pd.DataFrame(date)

# Extract the year from the 'date' column and create a new 'Year' column
date_dim['Year'] = date_dim['date'].dt.year

# Extract the Quarter from the 'date' column and create a new 'Quarter' column
date_dim['Quarter'] = date_dim['date'].dt.quarter

# Extract the month from the 'date' column and create a new 'Month' column
date_dim['Month'] = date_dim['date'].dt.month

# Extract the monthname  from the 'date' column and create a new 'Monthname' column
date_dim['Month_Name'] = date_dim['date'].dt.month_name()

# Extract the week from the 'date' column and create a new 'week' column
date_dim['Week'] = date_dim['date'].dt.weekday

# Extract the day from the 'date' column and create a new 'day' column
date_dim['day'] = date_dim['date'].dt.day

# Extract the day_name from the 'date' column and create a new 'day_name' column
date_dim['day_name'] = date_dim['date'].dt.day_name()

# Convert the 'date' column to just the date (without time) and update the 'date' column
date_dim['date'] = pd.to_datetime(date_dim['date']).dt.date
# Display the updated DataFrame
date_dim
```

## 2NF to 3NF
### Converting and Merging Account Table OpeningDates with Date Dimension Table

```python
# Changing the Account table for 2NF to 3NF
# Merge 'account' with 'date_dim' on 'OpeningDate' and 'date', rename 'date_id' to 'OpeningDate_ID',
# then select and keep the specified columns. 
# -----------------------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------------------------
# Convert 'OpeningDate' to just the date
account['OpeningDate'] = pd.to_datetime(account['OpeningDate']).dt.date

# Merge with date_dim on 'OpeningDate'
# left_on='OpeningDate': Specifies the column in the account DataFrame to match on (in this case, OpeningDate).
account = account.merge(date_dim, left_on='OpeningDate', right_on='date', how='inner') # 

# Rename 'date_id' to 'OpeningDate_ID'
account = account.rename(columns={'date_id': 'OpeningDate_ID'})

# Reset index and select only the needed columns
account = account.reset_index(drop=True)[['account_id', 'AccountType', 'Balance', 'OpeningDate_ID']]
```

### Converting and merging 'TranscationDates' on Transcation Table with Date Dimension Table

```python
# Convert 'TransactionDate' to just the date
transcations['TransactionDate'] = pd.to_datetime(transcations['TransactionDate']).dt.date

# Merge with date_dim on 'TransactionDate'
transcations = transcations.merge(date_dim, left_on='TransactionDate', right_on='date', how='inner')

# Rename 'date_id' to 'TransactionDate_ID'
transcations = transcations.rename(columns={'date_id': 'TransactionDate_ID'})

# Reset index and select only the needed columns
transcations = transcations.reset_index(drop=True)[['transcations_id', 'TransactionType', 'Amount','TransactionDate_ID']]
```

# Loan DWH FactTable
### Converting and Merging Loan Dates with Date Dimension Table

```python
# Convert 'StartDate' to just the date
# Convert 'StartDate' and 'EndDate' to just the date (without time)
loan['StartDate'] = pd.to_datetime(loan['StartDate']).dt.date
loan['EndDate'] = pd.to_datetime(loan['EndDate']).dt.date

# Merge with date_dim on 'StartDate' and rename 'date_id' to 'StartDate_ID'
loan = loan.merge(date_dim, left_on='StartDate', right_on='date', how='inner')
loan = loan.rename(columns={'date_id': 'StartDate_ID'})

# Merge with date_dim on 'EndDate' and rename 'date_id' to 'EndDate_ID'
loan = loan.merge(date_dim, left_on='EndDate', right_on='date', how='inner', suffixes=('', '_end'))
loan = loan.rename(columns={'date_id': 'EndDate_ID'})

# Reset index and select only the needed columns
loan = loan.reset_index(drop=True)[['loan_id', 'LoanType', 'LoanAmount', 'InterestRate', 'StartDate_ID', 'EndDate_ID']]
loan
```

### Loading Tables into the RDBMS using pyscopg2

```python
#  ! pip install psycopg2
import psycopg2
```

```python
import psycopg2

# Database parameters
db_params = {
    "username": "postgres",
    "password": "password",
    "host": "localhost",
    "port": 5432,
    "database": "Zulo_Bank_DB"
}

# Connect to the default "postgres" database to check/create the target database
db_url = "postgresql://{username}:{password}@{host}:{port}/postgres".format(**db_params)

try:
    conn = psycopg2.connect(db_url)
    conn.autocommit = True
    cursor = conn.cursor()

    # Check if the database exists
    cursor.execute(f"SELECT 1 FROM pg_database WHERE datname = '{db_params['database']}';")
    exists = cursor.fetchone()

    if not exists:
        cursor.execute(f'CREATE DATABASE "{db_params["database"]}";')
        print(f'Database "{db_params["database"]}" created successfully.')
    else:
        print(f'Database "{db_params["database"]}" already exists.')

    cursor.close()
    conn.close()
except Exception as e:
    print(f"Error: {e}")
```

#### Create a function to connect to the newly created database to create schema

```python
# Function to connect to the target database
def get_db_connection():
    db_url = "postgresql://{username}:{password}@{host}:{port}/{database}".format(**db_params)
    return psycopg2.connect(db_url)

try:
    conn = get_db_connection()
    conn.autocommit = True
    print("Connected to the target database successfully!")
except Exception as e:
    print(f"Error connecting to the target database: {e}")
    conn = None

if conn:
    try:
        cursor = conn.cursor()
        
        # Schema and table creation query
        create_table_query = '''




            CREATE SCHEMA IF NOT EXISTS zulobankdb;
            DROP TABLE IF EXISTS zulobankdb.date_dim CASCADE;
            DROP TABLE IF EXISTS zulobankdb.transactions CASCADE;
            DROP TABLE IF EXISTS zulobankdb.customer CASCADE;
            DROP TABLE IF EXISTS zulobankdb.account CASCADE;
            DROP TABLE IF EXISTS zulobankdb.loan CASCADE;


            CREATE TABLE IF NOT EXISTS zulobankdb.date_dim(
                date_id SERIAL PRIMARY KEY,
                date VARCHAR(100000),
                Year INTEGER,
                Month INTEGER,
                Quarter INTEGER,
                Month_Name VARCHAR(100),
                Week INTEGER,
                day INTEGER,
                day_name VARCHAR(1000)
            );

            CREATE TABLE IF NOT EXISTS zulobankdb.transactions(
                transactions_id SERIAL PRIMARY KEY,
                TransactionType VARCHAR(10000),
                Amount FLOAT,
                TransactionDate_ID INTEGER,
                FOREIGN KEY(TransactionDate_ID) REFERENCES zulobankdb.date_dim(date_id)
            );

            CREATE TABLE IF NOT EXISTS zulobankdb.customer(
                customer_id SERIAL PRIMARY KEY,
                first_name VARCHAR(10000),
                last_name VARCHAR(10000),
                Email VARCHAR(100000),
                Phone VARCHAR(10000000)
            );

            CREATE TABLE IF NOT EXISTS zulobankdb.account(
                account_id SERIAL PRIMARY KEY,
                AccountType VARCHAR(10000),
                Balance FLOAT,
                OpeningDate_ID INTEGER,
                FOREIGN KEY(OpeningDate_ID) REFERENCES zulobankdb.date_dim(date_id)
            );

            CREATE TABLE IF NOT EXISTS zulobankdb.loan(
                loan_id SERIAL PRIMARY KEY,
                LoanType VARCHAR(100000),
                LoanAmount FLOAT,
                StartDate_ID INTEGER,
                EndDate_ID INTEGER,
                FOREIGN KEY(StartDate_ID) REFERENCES zulobankdb.date_dim(date_id),
                FOREIGN KEY(EndDate_ID) REFERENCES zulobankdb.date_dim(date_id)
            );

            CREATE TABLE IF NOT EXISTS zulobankdb.zulo_bank_fact(
            customer_id INTEGER,
            account_id  INTEGER,
            transcations_id  INTEGER,
            loan_id INTEGER,
            FOREIGN KEY(customer_id) REFERENCES zulobankdb.customer(customer_id),
            FOREIGN KEY(account_id) REFERENCES zulobankdb.account(account_id),
            FOREIGN KEY(transcations_id) REFERENCES zulobankdb.transactions(transactions_id)

            )
        '''

        # Execute and commit the table creation query
        cursor.execute(create_table_query)
        print("Schema and tables created successfully.")
    except Exception as e:
        print(f"Error creating schema/tables: {e}")
    finally:
        cursor.close()
        conn.close()
```
